```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
test = sio.wavfile.read('data/cymbal44.wav', mmap=False)
#print(test)
cymbal = test[1][0:200000]
plt.plot(cymbal)
plt.savefig('figs/cymbal44.pdf')
plt.savefig('figs/cymbal44.eps')
```

```python
from importlib import reload
from math import sqrt, log2

import pytempscsp.tempscsp as tempscsp
import pytempscsp.tempscsel as tempscsel
```

```python
gamma = 1
c = sqrt(2)
sigmamin = 0.125
sigmamax = 16384.0
tempscspder1csqrt2, sigmas = tempscsel.limitkernders_mult(cymbal, sigmamin, sigmamax, 1, gamma, c)
amin = np.min(tempscspder1csqrt2)
amax = np.max(tempscspder1csqrt2)
arange = max(abs(amin), abs(amax))
plt.imshow(np.flipud(tempscspder1csqrt2), cmap='gray', aspect=9000, interpolation='nearest', vmin=-arange, vmax=arange, extent=[0, 200000, log2(sigmamin), log2(sigmamax)])
plt.set_cmap('bwr')
plt.tight_layout()
#plt.xticks(fontsize=14)
#plt.yticks(fontsize=14)
plt.savefig('figs/cymbal44-der1-csqrt2.pdf')
plt.savefig('figs/cymbal44-der1-csqrt2.eps')
```

```python
gamma = 1
c = sqrt(2)
sigmamin = 0.125
sigmamax = 16384.0
tempscspder2csqrt2, sigmas = tempscsel.limitkernders_mult(cymbal, sigmamin, sigmamax, 2, gamma, c)
amin = np.min(tempscspder2csqrt2)
amax = np.max(tempscspder2csqrt2)
arange = max(abs(amin), abs(amax))
plt.imshow(np.flipud(tempscspder2csqrt2), cmap='gray', aspect=9000, interpolation='nearest', vmin=-arange, vmax=arange, extent=[0, 200000, log2(sigmamin), log2(sigmamax)])
plt.set_cmap('bwr')
plt.tight_layout()
#plt.xticks(fontsize=14)
#plt.yticks(fontsize=14)
plt.savefig('figs/cymbal44-der2-csqrt2.pdf')
plt.savefig('figs/cymbal44-der2-csqrt2.eps')
```

```python
C = 1/sqrt(2)
quasi = np.sqrt(tempscspder1csqrt2**2 + C*tempscspder2csqrt2**2)
amin = np.min(quasi)
amax = np.max(quasi)
arange = max(abs(amin), abs(amax))
plt.imshow(np.flipud(quasi), cmap='gray', interpolation='nearest', aspect=9000, vmin=-arange, vmax=arange, extent=[0, 200000, log2(sigmamin), log2(sigmamax)])
plt.set_cmap('bwr')
plt.tight_layout()
#plt.xticks(fontsize=14)
#plt.yticks(fontsize=14)
plt.savefig('figs/cymbal44-quasi-csqrt2.pdf')
plt.savefig('figs/cymbal44-quasi-csqrt2.eps')
```
