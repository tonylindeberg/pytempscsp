```python
import tempscsp
import importlib
importlib.reload(tempscsp)
import matplotlib.pyplot as plt
from pyscsp.discscsp import variance1D

from math import sqrt
import numpy as np

from scipy.ndimage import correlate1d, correlate
```

```python
deltafcn = tempscsp.deltafcn1D(25)
dt = np.array([0,-1, 1])
dtt = np.array([0, 0, 1, -2, 1])
```
