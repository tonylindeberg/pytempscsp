```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt

vib = np.loadtxt("data/2003.11.25.23.39.56")[:, 1] + 0.1

vib = vib[1:2000]
plt.plot(vib)
plt.savefig('figs/vib.pdf')
plt.savefig('figs/vib.eps')
```

```python
from math import sqrt, log2
from importlib import reload

import pytempscsp.tempscsp as tempscsp
import pytempscsp.tempscsel as tempscsel
```

```python
gamma = 1
c = sqrt(2)
sigmamin = 0.125
sigmamax = 1024.0
tempscspder1csqrt2, sigmas = tempscsel.limitkernders_mult(vib, sigmamin, sigmamax, 1, gamma, c)
amin = np.min(tempscspder1csqrt2)
amax = np.max(tempscspder1csqrt2)
arange = max(abs(amin), abs(amax))
plt.imshow(np.flipud(tempscspder1csqrt2), cmap='gray', aspect=100, interpolation='nearest', vmin=-arange, vmax=arange, extent=[0, 2000, log2(sigmamin), log2(sigmamax)])
plt.set_cmap('bwr')
plt.tight_layout()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('figs/vib-der1-csqrt2.pdf')
plt.savefig('figs/vib-der1-csqrt2.eps')
```

```python
gamma = 1
c = sqrt(2)
sigmamin = 0.125
sigmamax = 1024.0
tempscspder2csqrt2, sigmas = tempscsel.limitkernders_mult(vib, sigmamin, sigmamax, 2, gamma, c)
amin = np.min(tempscspder1csqrt2)
amax = np.max(tempscspder1csqrt2)
arange = max(abs(amin), abs(amax))
plt.imshow(np.flipud(tempscspder2csqrt2), cmap='gray', aspect=100, interpolation='nearest', vmin=-arange, vmax=arange, extent=[0, 2000, log2(sigmamin), log2(sigmamax)])
plt.set_cmap('bwr')
plt.tight_layout()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('figs/vib-der2-csqrt2.pdf')
plt.savefig('figs/vib-der2-csqrt2.eps')
```

```python
C = 1/sqrt(2)
quasi = np.sqrt(tempscspder1csqrt2**2 + C*tempscspder2csqrt2**2)
amin = np.min(quasi)
amax = np.max(quasi)
arange = max(abs(amin), abs(amax))
plt.imshow(np.flipud(quasi), cmap='gray', interpolation='nearest', aspect=100, vmin=-arange, vmax=arange, extent=[0, 2000, log2(sigmamin), log2(sigmamax)])
plt.set_cmap('bwr')
plt.tight_layout()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('figs/vib-quasi-csqrt2.pdf')
plt.savefig('figs/vib-quasi-csqrt2.eps')
```
