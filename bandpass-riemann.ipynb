```python
# The test signal 'Riemann', originally due to Donoho, generated by the function pywt.data.demo_signal() for in the PyWavelets library for n = 30.
riemann = pywt.data.demo_signal(name='Riemann', n=30)
```

```python
plt.plot(riemann)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.gca().set_ylim([0, 0.08])
plt.gca().axes.set_box_aspect(0.3)
plt.tight_layout() 
plt.savefig('figs/riemann.pdf', transparent=True)
plt.savefig('figs/riemann.eps', transparent=True)
```

```python
(riemannsmooth, scales) = tempscsp.limitkernfilt_mult(riemann, 1, 8, 2, 4)
```
