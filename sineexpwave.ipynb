```python
from math import sqrt, log2
import numpy as np

from importlib import reload
import matplotlib.pyplot as plt

from pyscsp.discscsp import variance1D
import pytempscsp.tempscsp as tempscsp
import pytempscsp.tempscsel as tempscsel
```

```python
reload(tempscsel)
```

```python
t = np.linspace(1,1000,1000)
a = 200
b = 1000
sine = np.sin(np.exp((b-t)/a))
plt.plot(sine)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('figs/sineexpwave.pdf')
plt.savefig('figs/sineexpwave.eps')
```

```python
t = np.linspace(1,1000,1000)
a = 200
b = 1000
model = np.sin(np.exp((b-t)/a))
gamma = 1
c = sqrt(2)
sigmamin = 0.125
sigmamax = 1024.0
tempscspder1csqrt2, sigmas = tempscsel.limitkernders_mult(model, sigmamin, sigmamax, 1, gamma, c)
amin = np.min(tempscspder1csqrt2)
amax = np.max(tempscspder1csqrt2)
arange = max(abs(amin), abs(amax))
plt.imshow(np.flipud(tempscspder1csqrt2), cmap='gray', aspect=50, interpolation='nearest', vmin=-arange, vmax=arange, extent=[0, 1000, log2(sigmamin), log2(sigmamax)])
plt.set_cmap('bwr')
plt.tight_layout()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('figs/sineexpwave-der1-csqrt2.pdf')
plt.savefig('figs/sineexpwave-der1-csqrt2.eps')
```

```python
t = np.linspace(1,1000,1000)
a = 200
b = 1000
model = np.sin(np.exp((b-t)/a))
gamma = 1
c = sqrt(2)
sigmamin = 0.125
sigmamax = 1024.0
tempscspder2csqrt2, sigmas = tempscsel.limitkernders_mult(model, sigmamin, sigmamax, 2, gamma, c)
amin = np.min(tempscspder2csqrt2)
amax = np.max(tempscspder2csqrt2)
arange = max(abs(amin), abs(amax))
plt.imshow(np.flipud(tempscspder2csqrt2), cmap='gray', interpolation='nearest', aspect=50, vmin=-arange, vmax=arange, extent=[0, 1000, log2(sigmamin), log2(sigmamax)])
plt.set_cmap('bwr')
plt.tight_layout()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('figs/sineexpwave-der2-csqrt2.pdf')
plt.savefig('figs/sineexpwave-der2-csqrt2.eps')
```

```python
C = 1/sqrt(2)
quasi = np.sqrt(tempscspder1csqrt2**2 + C*tempscspder2csqrt2**2)
amin = np.min(quasi)
amax = np.max(quasi)
arange = max(abs(amin), abs(amax))
plt.imshow(np.flipud(quasi), cmap='gray', interpolation='nearest', aspect=50, vmin=-arange, vmax=arange, extent=[0, 1000, log2(sigmamin), log2(sigmamax)])
plt.set_cmap('bwr')
plt.tight_layout()
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.savefig('figs/sineexpwave-quasi-csqrt2.pdf')
plt.savefig('figs/sineexpwave-quasi-csqrt2.eps')
```
